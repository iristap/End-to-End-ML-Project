```python
from sklearn.model_selection import train_test_split
```

```python
df.head()
```

```python
X = df.drop(['quality'], axis=1) # features 
y = df['quality'] # target
```

```python
X.head()
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=10203040) 
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import ElasticNet # regression model
# 
```

```python
lr = ElasticNet(alpha=0.1, l1_ratio=0.5, random_state=77)
lr.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import numpy as np
```

```python
def eval_metrics(actual, pred):
    rmse = np.sqrt(mean_squared_error(actual, pred))
    mae = mean_absolute_error(actual, pred)
    r2 = r2_score(actual, pred)

    return rmse, mae, r2
```

```python
predicted_value = lr.predict(X_test)
```

```python
predicted_value
```

```python
(rmse, mae, r2) = eval_metrics(y_test, predicted_value)
```

```python
print("rmse = ", rmse)
print("mae = ", mae)
print("r2 = ", r2)
```
